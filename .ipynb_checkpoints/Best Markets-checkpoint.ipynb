# Finding the Best Markets to Advertise in for E-Learning

In this project, we will act as if we work for an e-learning company offering courses in a wide variety of programming areas, but focusing on web and mobile development. Our goal is to promote our product by determining the two best markets to advertise in.

## Digging Into the Data

Since organizing and implementing surveys can be costly and time-intensive, we'll be using a dataset from 
[freecodeCamp's 2017 New Coder Survey](https://medium.freecodecamp.org/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969). 

Because the company has a publication on Medium that attracts new coders with a variety of articles not limited to *just* web/mobile development, the survey data we analyze should be fairly representative of the interests of the new coder population as a whole.


Below, we'll read in and explore the data from the [public GitHub repo](https://github.com/freeCodeCamp/2018-new-coder-survey)

```python
#import libraries we'll need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#change way notebook displays tables/charts
%matplotlib inline
pd.options.display.max_columns = 150
```

```python
#read in data
survey_data = pd.read_csv('2017-fCC-New-Coders-Survey-Data.csv', low_memory = 0) #low_memory to remove dtype warnings

#quick data exploration
print('rows = {}, columns = {}'.format(survey_data.shape[0],survey_data.shape[1]))
survey_data.head(5)
```

```python
list(survey_data.columns)
```

### Describing the Dataset:
Checking the columns against the [Medium Post](https://www.freecodecamp.org/news/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969/), we can see that the columns can be broken down into roughly three categories:
- platforms used for learning (YouTube, CodeEvent, Bootcamp, etc.)
- Job interests (JobInterest)
- demographics (Has, Is, Employment, etc.)

## Checking Sample Representivity
In order to determine where to spend our ad dollars, we want to know:

* where new coders are located
* which places have the highest density of new coders
* how much money they're willing to spend on learning

First, we need check that this sample of is representative of the population we are attempting to advertise to, namely, new coders who are interested in web or mobile development.

As a proxy for identifying respondents interests, we'll assume that people who are interested in *finding a job* in a certain field are also interested in *learning* about that field. To this end, we'll begin by investigating the **`JobRoleInterest`** column for the types of jobs respondents are intereseted in

```python
jobinterests = survey_data['JobRoleInterest'].value_counts(normalize=True)*100
jobinterests.head(10)
```

### Understanding New Coders Interests
Looking at the output above, we can see that based on the distribution of the top responses (the top 5 responses make up only 20% of the distribution) and the granularity of the responses (multiple jobs in responses towards bottom), it appears many of the respondents have multiple interests. 

To determine how many of them have singular vs. multiple interests, we will:

- remove null values in `JobInterests` (since we can't apply string operations to those)
- split each entry by the `Series.str.split(',')` method
- count the length of the resulting lists
- create a frequency table of the lengths of those lists

```python
#split string in job interest column
jobs_no_nulls = survey_data['JobRoleInterest'].dropna()
splitted_interests = jobs_no_nulls.str.split(',')

#count length of string, output results
n_of_options = splitted_interests.apply(lambda x: len(x))
n_of_options.value_counts(normalize=True)*100
```

It appears as if only 31% of the respondents know exactly what type of job they want, with the rest having multiple job role interests.
Let's identify how many respondents are interested in our core focus, mobile and web development.

```python
#create frequency table for rows containing Mobile/web
patt= 'Web Developer|Mobile Developer'
webormobile = jobs_no_nulls.str.contains(patt) 
freq_table = webormobile.value_counts(normalize=True) * 100
print(freq_table)

#create graph 
plt.style.use('fivethirtyeight')
bargraph = freq_table.plot.bar(color=['blue','red'])
plt.title('Most Respondents are Interested in \nWeb or Mobile Development')
plt.ylabel('Percentage', fontsize=12)
plt.xticks([0,1],['Web or Mobile','Other Subject'],rotation=0) #original xtick labels were True, False
plt.ylim(0,100)
plt.show()
```

From the above figure, over 85% of the respondents are interested in mobile and web development. This is a strong argument for considering this sample representative of our population of interest: new coders interested in web or mobile development.

We've confirmed the sample set is representative of our target audience. To figure out where we want to advertise, we need to determine:
- where new coders are located
- which locations have the greatest number of new coders
- how much money new coders are willing to spend on learning

## Which Countries Have the Most New Coders?
Let's determine where new coders are located. We'll first look at frequency table for the `CountryLive` column, the country where each respondent is currently residing. To ensure we're only looking at coders who meet our proxy criteria for "looking to learn", like above, we'll remove all respondents that didn't respond to the `JobRoleInterest` column

```python
rep_survey = survey_data[survey_data['JobRoleInterest'].notnull()].copy()
abs_freq = rep_survey['CountryLive'].value_counts()
rel_freq = rep_survey['CountryLive'].value_counts(normalize=True)*100

pd.DataFrame(data = {'Absolute Frequency':abs_freq,
                    'Percentage': round(rel_freq, 2)})
```

45.7% of potential customers are in the US which should be our main focus. After that, India has about 8%, roughly equivalent to the combined rates of the next two countries,  Canda and the UK, combined. Luckily these are all countries with English speakers. But, how much each country's respondents have to spend on learning needs to be taken into account for our advertising needs as well.


## Which Countries Spend the Most on Learning?
Determining how much money our population is willing to spend (since we are selling a paid service) is a key factor in deciding where we want to buy adspace, equally important as finding the largest market. Next we'll identify how much each of our top countries spend on average, per month on elearning by:

- creating `money_per_month` col: the division of `MoneyForLearning` by `months programming`
- removing all the null values from our new column
- exploring the range of values each of our top countries spend
- getting an accurate average for the amount spent

```python
#replace 0s with 1s to avoid dividing by 0
rep_survey['MonthsProgramming'].replace(0,1, inplace=True)

#create new column
rep_survey['money_per_month'] = rep_survey['MoneyForLearning'] / rep_survey['MonthsProgramming']
print('of {} individuals, {} contain null values in `money_per_month`'.format(rep_survey.shape[0], rep_survey['money_per_month'].isna().sum()))
```

```python
#keep only rows w/ non-null values in Country, money per month for our analysis
rep_survey = rep_survey[rep_survey['money_per_month'].notnull()]
rep_survey = rep_survey[rep_survey['CountryLive'].notnull()]
```

```python
#check to see that we still have enough values to make a determination
new_abs_freq = rep_survey['CountryLive'].value_counts()
sample_reduction =  new_abs_freq.head() / abs_freq.head()
print(sample_reduction)
```

We still have over 85% of our data for each of the top countries, which should be large enough to analyze with.
Let's compute the average each country's repsondent spends per month.

```python
survey_by_country = rep_survey.groupby('CountryLive')
survey_by_country['money_per_month'].agg(np.mean)[['United States of America', 'Canada', 'United Kingdom', 'India']]
```

## Dealing With Outliers and Bootcamps

From our average spent table above, India appears to spend more than *either* Canada or the UK, which would be surprising comparing this metric to say, GDP per capita of the respective nations, especially considering how little the UK seems to spend in comparison. Let's check the values to make sure that there isn't any outliers skewing our calculated values. Below, we'll make a box plot to see how much each respondent spends per month

```python
#select top countries
target_countries = rep_survey[rep_survey['CountryLive'].str.contains('United States of America|India|United Kingdom|Canada')]

#create boxplot
sns.boxplot(x='CountryLive',y='money_per_month', data=target_countries)
#labeling
plt.title('Money Spent Per Month Per Country',
          y=1.02) #pads title up slightly
plt.xlabel('Country',fontsize=14)
plt.ylabel('Money Spent Per Month',fontsize=14)
plt.xticks(range(4), ['US','UK','India','Canada']) #avoid lengthy ticklabels overlap
plt.show()
```

The scale of the plot is skewed by a couple outliers where US respondents are spending more than  $\$$50000 per month. Since these are such large outliers, we'll remove them from our analysis

```python
#Isolate participants who spend less than 20000 per month
rep_survey = rep_survey[rep_survey['money_per_month'] < 20000]
```

```python
#recompute means
countries_mean = rep_survey.groupby('CountryLive').mean()
countries_mean['money_per_month'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
#recalculate and make new table
target_countries = rep_survey[rep_survey['CountryLive'].str.contains('United States of America|India|United Kingdom|Canada')]

sns.boxplot(x='CountryLive',y='money_per_month', data=target_countries)
plt.title('Money Spent Per Month Per Country',
          y=1.02) 
plt.xlabel('Country',fontsize=14)
plt.ylabel('Money Spent Per Month (USD)',fontsize=14)
plt.xticks(range(4), ['US','UK','India','Canada'], fontsize=11) 
plt.show()
```

From our new plot, it looks like there are a few values in India and Canada that are outliers in their respective distributions. This could be from respnodents attending bootcamps, which are more expensive than online learning. Let's investigate how common bootcamps are for high spenders.

Strategy: Remove those who have spent money on bootcamps but have less than 3 months of programming, as this is a high initial investment that is skewing our data
Remove outliers from each country

```python
arbval = 2500
```

```python
rich_India = target_countries[((target_countries['CountryLive'] == 'India') & (target_countries['money_per_month'] >= arbval))]
print('India: {} % spending above {} attended bootcamp'.format(round(rich_India['AttendedBootcamp'].sum() / rich_India.shape[0]*100, 2), arbval))

rich_India[['AttendedBootcamp', 'money_per_month', 'MonthsProgramming']].sort_values(['money_per_month'],ascending = False)
```

None of the high spending respondents in India attendeded a bootcamp. The question from the survey asks how much money *aside* from univeristy tuition was spent on learning to code, so maybe these respondents misunderstood the question and included university tuition. In any case, Since it's so difficult to determine *how* they spent so much money without attending a bootcamp, we'll remove these rows.

```python
#drop India outliers
target_countries = target_countries.drop(rich_India.index) #use index to drop
```

Looking back at the box plot above, there are still a number of extreme outliers in the US (above $\$$6000 per month) in comparison with the rest of the distribution. Let's see 

```python
rich_US = target_countries[((target_countries['CountryLive'] == 'United States of America') & (target_countries['money_per_month'] > 6000))]
print('US: {} % spending above {} attended bootcamp'.format(round(rich_US['AttendedBootcamp'].sum() / rich_US.shape[0]*100, 2), 6000))
rich_US[['AttendedBootcamp', 'money_per_month', 'MonthsProgramming']].sort_values(['money_per_month','MonthsProgramming'],ascending=False)
```

With US respondents, six out of the eleven extreme outliers had taken a bootcamp. Again, it's hard to determine from the data where non-bootcamp attendees would have spent this much on learning, so we'll remove those that hadn't atteneded bootcamps from our analysis.

There's something else. Eight of the respondents hadn't been programming for more than three months. Because a bootcamp is a large upfront investment and lasts for several months, even though these individuals may have spent a large sum at the time of the survey taking place, they may not have finished their bootcamps by this point. The additional, future months that large cost will ultimately be spread across means the actual cost per month is more than it ought to. For this reason, we'll remove these outliers as well.

Below we'll remove respondents that:
* didn't attend bootcamps
* had been programming for $\leq$ 3 months

```python
#remove outlying US respondents who hadn't attended bootcamp
no_bootcamp_or_novices = rich_US[(rich_US['AttendedBootcamp'] == 0) | 
                                 (rich_US['MonthsProgramming'] <= 3)]
target_countries.drop(no_bootcamp_or_novices.index,inplace=True) #remove via row indices
```

From the above boxplot, there appears to be one Canadian outlier spending $\$$5000 per month, let's investigate.

```python
rich_CA = target_countries[((target_countries['CountryLive'] == 'Canada') & (target_countries['money_per_month'] > 4500))]
rich_CA[['AttendedBootcamp', 'money_per_month', 'MonthsProgramming']]
```

Since this person falls into the category of "new bootcamp attendee" (< 3 months programming) we developed in the US, we'll remove this datapoint as well.

```python
#remove outlying CA respondent 
target_countries.drop(rich_CA.index,inplace=True)
```

```python
sns.boxplot(x='CountryLive',y='money_per_month', data=target_countries)
plt.title('Money Spent Per Month Per Country',
          y=1.02) #pads title up slightly
plt.xlabel('Country',fontsize=12)
plt.ylabel('Money Spent Per Month',fontsize=14)
plt.xticks(range(4), ['US','UK','India','Canada']) #avoid ticklabels overlap
plt.show()
```

## Picking Two Markets
Without a doubt, **the US is a market we want to advertise in**, probably the primary market we want to advertise in. By the numbers, it's the home of 75% of our top four countries, roughly 50% of all respondents, and new coders there are willing to spend $\$$143 per month on learning. Our real problem is picking the second country we wish to advertise in.

Our subscription sells for $\$$59 per month. Based on dollars spent per month, Canada appears to be a solid second choice with an average of $\$$93 dollars spent per month. However, people in India also meet our $\$$59 criteria, coming in at $\$$65 spent per month, and also have roughly twice as many potential customers as in Canada:

```python
#Frequency Table for target countries 'CountryLive' column
target_countries['CountryLive'].value_counts(normalize=True) * 100
```

Picking the second country to advertise in becomes even murkier when we consider whether we should even include bootcamp respondents in our analysis.

```python
#filter out bootcamp respondents
no_boot_target_countries = target_countries[target_countries['AttendedBootcamp'] == 0]
print('Average Spending per Country: incl. Bootcamps\n {}\n'.format(
    target_countries.groupby('CountryLive')['money_per_month'].agg(np.mean).sort_values(ascending=False)))
print('Average Spending per Country: No Bootcamps\n {}\n'.format(
    no_boot_target_countries.groupby('CountryLive')['money_per_month'].agg(np.mean).sort_values(ascending=False)))
```

```python
#percentage of respondents who had attended Bootcamp
(target_countries['AttendedBootcamp'] == True).sum() / target_countries.shape[0]
```

### Should We Count Bootcamps?

When we examine the frequency tables of money spent per month by country excluding bootcamps, the results are surprising. US spending is cut in half, and the gap between spending in Canada and India is all but eliminated, making India appear the better option based on potential customers.

Now, this is all dependent on *not* counting people who are attending bootcamps in our analysis. With programming, the goal of any coursework is posited here to be helping people to learn skills that can land them jobs, promotions, etc. Many bootcamps explicitly (or implicitly) promise participants they will be "job ready" upon completion of the course, at which point they would ostensibly be learning on the job instead of online.

This brings us to the following line of proxy questions:

- "Are people who attend bootcamps in our target audience?" => 
- "Do people who attend bootcamps also pay for online learning?" => 
- **"Do people who attend bootcamps find jobs soon afterwards?"**

Considering the disparity, the answer to this question sets a strong foundation for future analysis on the subject.

### Final Determination

In any case, our options for where to market are in some proportion the following:
1. Advertise in the US and Canada *or* the US and India
2. Advertise only in the US.
3. Advertise in the US, India, and Canada.

**We suggest sending our results to the marketing department**, who's domain expertise will help make a more clear determination of where and how to spend by taking into account factors such as cultural disparities between US/India that could adversely drive up advertisement development costs.
